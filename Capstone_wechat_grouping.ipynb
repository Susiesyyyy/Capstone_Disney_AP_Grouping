```python
with open('AP sample data 1101 0131/AP Expire header.txt') as f:
    lines = f.read().splitlines() 
    line = lines[0]
    
header = line.split("\t")

active = pd.read_csv('AP sample data 1101 0131/AP Expired_20211101_20220131.csv')
active.columns = header
active.head()
```

```python
multiple_wechat_binding = active['binding_wechat_open_id_num'].value_counts().to_frame()
multiple_wechat_binding = multiple_wechat_binding[multiple_wechat_binding['binding_wechat_open_id_num']>1]
multiple_wechat_binding
```

```python
wechat_group = active.groupby('binding_wechat_open_id_num')['source_id'].apply(list)
wechat_group = wechat_group[wechat_group.str.len() > 1][wechat_group.str.len() < 10].reset_index(drop=True)
wechat_group = wechat_group.to_list()#.to_csv('wechat_group.csv')
```
